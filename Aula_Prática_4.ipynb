```python
def esti_raiz(vet_coef):
  qv = 0 # Coeficiente de quantos raizes pode ter
  i = 0 # Parametro para loop
  j = 1 # Paramentro para loop

  #encontra coeficiente para encontrar raizes positivas
  while i<len(vet_coef)-1 and j < len(vet_coef):
    if vet_coef[i]*vet_coef[j]< 0:
      qv= qv + 1
      i = i + 1
      j = j + 1
    elif vet_coef[i]*vet_coef[j] == 0:
      if vet_coef[j] == 0:
        j = j + 1
      else:
        i = i +1
    else:
      i = i +1
      j = j + 1

  # numero de raizes reais positivas ()
  nrp = []
  for i in range(0,qv+1):
    if(qv - i)%2 == 0:
      nrp.append(i)

  # reiniciando as variaveis
  qv = 0
  i= 0
  j = 1

  #Mudar o que tem grau impar para poder terstar as raizes negativas
  for k in range(0,len(vet_coef)):
    vet_coef[k] = vet_coef[k]*(-1**k)

  #Repetindo o processo para calculo de raizes negativas
  while i<len(vet_coef)-1 and j < len(vet_coef):
    if vet_coef[i]*vet_coef[j]< 0:
      qv= qv + 1
      i = i + 1
      j = j + 1
    elif vet_coef[i]*vet_coef[j] == 0:
      if vet_coef[j] == 0:
        j = j + 1
      else:
        i = i + 1
    else:
      i = i + 1
      j = j + 1

  #Numero de raizes negativas ()
  nrn = []
  for i in range (0,qv+1):
    if(qv - i)%2 ==0:
      nrn.append(i)
  #Numero de raizes complexas ()
  nrc = []
  grau = len(vet_coef)-1
  for i in nrp:
    for j in nrn:
      nrc.append(grau - i - j)

  return nrp,nrn, nrc

vet_coef = [1,1,0,-4,-3,2]
[nrp,nrn,nrc] = esti_raiz(vet_coef)
print(nrp,nrn,nrc)
```

```python
def calc_pool_conv(vetor,grau,x):
  soma = vetor[0]
  for i in range(1,grau+1):
    soma = soma + vetor[i]*(x**i)
  return soma

def metodo_horner(vetor,grau,x):
  soma = vetor[grau]
  for i in range(0,grau):
    soma = soma*x + vetor[grau-i-1]
  return soma

y = calc_pool_conv(vet_coef,5,2)
z = metodo_horner(vet_coef,5,2)

print(y)

print(z)

```
