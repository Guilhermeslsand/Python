```python
#Questão 1
import matplotlib.pyplot as plt
import numpy as np
import math

def f(x):
  return 5*(x**3)-5*(x**2)+6*x-2

x=np.linspace(0.4,0.6,1000)
y=[]
for i in x:
  y.append(f(i))
plt.grid()
plt.plot(x,y)

def bisseccao(a,b,tol,maxiter):
  fa=f(a)
  fb=f(b)
  xr=(a+b)/2
  fxr=f(xr)
  erro = 1e6
  it = 1
  while it < maxiter and erro > tol:
    if fa*fxr<0:
      b=xr
      fb=fxr
    elif fb*fxr < 0:
      a = xr
      fa = fxr
    elif fxr == 0:
      return 0
    xra = xr
    xr = (a+b)/2
    fxr = f(xr)
    erro = math.fabs(xr-xra)/math.fabs(xr)
    it = it+1
  return (xr,it,erro)

print(bisseccao(0,1,0.1,100))
```
